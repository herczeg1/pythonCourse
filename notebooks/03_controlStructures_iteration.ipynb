```python
#given a number with comas print only the numbers
number = '1,234,583,098,675,247'
#output: '1234583098675247'


def numberize(num):
    for i in range(0, len(num)):
        if num[i] in '0123456789':
            #print(num[i])
            print(num[i],end='')
numberize(number)
```

```python
#with a return value
def numberize_2(num):
    clean = ''
    for i in range(0, len(num)):
        if num[i] in '0123456789':
            #print(num[i])
            clean += num[i]
    return int(clean)  #look where we are with the indentation

print(numberize_2(number))
```

```python
# create a program that takes an IP address antered at the keyboard and prints out the number
#of segments it contains and the length og each segment

# 127.0.0.1
# 192.168.0.23
# 213.21.0.9

def analyze_ip():
    ip_add = input("Give me an IP: ")
    octet_lengths = [0,0,0,0]
    octet_position = 1
    segment_length=0
    total_length = len(ip_add)
    current_length = 1
    
    for char in ip_add:
        total_length = len(ip_add)
        if octet_position > 4:
            print('Too many octets.')
            break
            
        if char in "0123456789":
            segment_length +=1
            if (current_length == total_length): 
                octet_lengths[octet_position - 1] = segment_length
        elif char == '.':
            octet_lengths[octet_position - 1] = segment_length
            segment_length = 0
            octet_position +=1
        current_length += 1
    return(octet_lengths)

analyze_ip()
```
